```python
from faker import Faker
import pandas as pd
import random
```

```python
fake = Faker()
```

```python
print(fake.name())
print(fake.email())
print(fake.company())
```

```python
def generate_subscription_data(num_rows=1000):
    data = []
    plans = ['Basic', 'Pro', 'Enterprise']
    
    for i in range(num_rows):
        sign_up_date = fake.date_between(start_date='-2y', end_date='today')
        
        record = {
            'Customer_ID': i + 1001,
            'Name': fake.name(),
            'Email': fake.email(),
            'Country': fake.country(),
            # FIXED: Pick from the plans list instead of a random number
            'Subscription_plan': random.choice(plans), 
            'Sign_up_date': sign_up_date,
            'Monthly_rate': round(random.uniform(9.99, 49.99), 2),
            'Is_Active': fake.boolean(chance_of_getting_true=75),
            'Last_login': fake.date_between(start_date=sign_up_date, end_date='today'),
            'Support_tickets': random.randint(0, 10)
        }
        data.append(record)
        
    # FIXED: Move return outside the for loop
    return pd.DataFrame(data)

df = generate_subscription_data(500)
print(df.head())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\ADMIN\Documents\Subscription_data_cleaned_csv.csv")
```

```python
df.columns
```

```python
numeric_df = df[['Monthly_rate','Is_Active','Support_tickets']]
correlation_matrix = numeric_df.corr()
```

```python
print(correlation_matrix['Is_Active'].sort_values(ascending=False))
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix, annot=True, cmap='RdYlGn', center = 0)
plt.title('Customer churn Correlation Heatmap')
plt.show()
```
